# MODEL BUILDING Churn Prediction in Telecom Industry

```python
import pandas as pd
```

```python
tel_dum=pd.read_csv('tel_customer_dummies.csv')
tel_dum
```

```python
tel_dum=tel_dum.drop(columns=['Unnamed: 0'])
```

```python
tel_dum
```

```python
y=tel_dum['Churn'].values
```

```python
x=tel_dum.drop(columns=['Churn'])
```

```python
x
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
feature=x.columns.values
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
scaler.fit(x)
```

```python
print(scaler.data_max_)
```

```python
print(scaler.data_min_)
```

```python
scaler.transform(x)
```

```python
x=pd.DataFrame(scaler.transform(x))
x.columns=feature
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
result=lr.fit(x_train,y_train)
```

```python
from sklearn import metrics
test=lr.predict(x_test)
print(metrics.accuracy_score(y_test,test))
```

Here is the model accuracy score 

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, test, labels=[0,1]))
```

Determine any null value is present or not

```python
tel_customer1=tel_customer
```

```python
tel_customer1.TotalCharges=pd.to_numeric(tel_customer1.TotalCharges,errors='coerce')
tel_customer1.isnull().sum()
```

Convert TotalCharges into numeric to find is there any null value is present or not. After applying to_numeric() and check there
is any null value or not. Here we determine there are 11 null values.

```python
tel_customer1[tel_customer1['TotalCharges'].isnull()==True]
```

```python
tel_customer1.dropna(inplace=True)
```

```python
tel_customer1['tenure'].max()
```

```python
label=["{0}-{1}".format(i,i+11) for i in range(1,72,12)]
tel_customer1['tenure_group'] = pd.cut(tel_customer1.tenure, range(1, 80, 12), right=False, labels=label)
```

```python
tel_customer1['tenure_group'].value_counts()
```

```python
tel_customer1.drop(columns=['customerID','tenure'],inplace=True,axis=1)
tel_customer1.head()
```

# Univariate Analysis

```python
for i, predictor in enumerate(tel_customer1.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=tel_customer1, x=predictor, hue='Churn')
```

```python
tel_customer1['Churn']=np.where(tel_customer['Churn']=='Yes',1,0)
```

Convert Yes into '1' and no into '0'

```python
tel_customer1.head()
```

```python
tel_customer_dummies=pd.get_dummies(tel_customer1)
tel_customer_dummies.head()
```

```python
plt.scatter(x=tel_customer1.MonthlyCharges,y=tel_customer1.TotalCharges)
plt.title('Scatter plot to identify relation between Monthly Charges and Total Charges')
plt.show()
```

# Determine the Monthly Charges of Churn and Non-Churn Customer

```python
sns.kdeplot(tel_customer_dummies.MonthlyCharges[tel_customer_dummies['Churn']==1],color='red')
sns.kdeplot(tel_customer_dummies.MonthlyCharges[tel_customer_dummies['Churn']==0],color='orange')
plt.legend(['Churn Yes','Churn No'],loc='upper right')
plt.title('Monthly Charges by Churn')
plt.show()
```

# Determine the Total Charges of Churn and Non-Churn Customers

```python
sns.kdeplot(tel_customer_dummies.TotalCharges[tel_customer_dummies['Churn']==1],color='red')
sns.kdeplot(tel_customer_dummies.TotalCharges[tel_customer_dummies['Churn']==0],color='orange')
plt.legend(['Churn Yes','Churn No'],loc='upper right')
plt.title('Total Charges by Churn')
plt.show()
```
